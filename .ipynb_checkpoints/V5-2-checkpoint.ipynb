```python
# imports
import pandas as pd
import numpy as np
from tkinter import *
from tkinter import filedialog
from tkinter import messagebox
from tkinter import ttk
import csv
```

```python
#Create a function to select a file
def fileOpen():
    fileName = filedialog.askopenfilename(initialdir = "/",
                                               title = "select a file", 
                                               filetypes=(("All files", "*.*"),))

    if fileName:
        try:
            global df
            fileName = r"{}".format(fileName)
            df = pd.read_excel(fileName, sheet_name='Main Areas required')      
            createTree()
            output_text('If this data looks correct, click the execute button')
        except:
            output_text('No data found, please try again.')
```

```python
#Create a function to change the output text label 
def output_text(text):
    
    my_label["text"] = text
```

```python
#Function to clear the preview screen
def clearTree():
    my_tree.delete(*my_tree.get_children())
```

```python
#function to perform the changes
def execute():
    try:
        #Create new table 
        
        global df
        df = df[['Event Name or interview','First name', 'Last name', 'Email', 'Date of birth updated ',
                    'Hubspot Country of residence (Domicile)', 'Audience Type',
               'Persona Type', 'Lead Type', 'Contact Type', 'Intended Year of Entry',
               'Postal code', 'Hubspot Subject of interest',
                    'Hubspot Subject area of interest - other']]

        #format DOB
        df['DOB'] = df["Date of birth updated "].dt.strftime('%d/%m/%Y')

        df = df[['Event Name or interview','First name', 'Last name', 'Email', 'DOB', 'Hubspot Country of residence (Domicile)',
                    'Audience Type', 'Persona Type', 'Lead Type', 'Contact Type', 'Intended Year of Entry',
                 'Postal code','Hubspot Subject of interest', 'Hubspot Subject area of interest - other']]

        # clean data and format table

        #New headers
        df = df.rename(columns = {'Event Name or interview': 'Name of event or interview',
                        'DOB': 'Date of birth updated',
                        'Hubspot Country of residence (Domicile)': 'Country of residence (Domicile)',
                         'Hubspot Subject of interest':'Subject of interest',
                          'Hubspot Subject area of interest - other': 'Subject area of interest - other'})


        #Add data to Audience Type
        df.loc[(df['Country of residence (Domicile)'] == 'United Kingdom') & (df['Audience Type'].isnull()), 'Audience Type']  = 'Home'

        df.loc[(df['Country of residence (Domicile)'] != 'United Kingdom') & (df['Country of residence (Domicile)'].notnull()) & (df['Audience Type'].isnull()), 'Audience Type'] = 'International'

        #Add data to persona type
        df.loc[df['Persona Type'].isnull(), 'Persona Type'] = 'Prospect Student'

        #Add data to Lead Type
        df.loc[df['Lead Type'].isnull(), 'Lead Type'] = 'Undergraduate'

        #Add data to contact type
        df.loc[df['Lead Type'] == 'Undergraduate', 'Contact Type'] = 'UG'
        df.loc[df['Lead Type'] != 'Undergraduate', 'Contact Type'] = 'PG'

        #remove rows with no data

        df.dropna(axis= 'index', how='any', subset = ['Last name','Email'], inplace = True)

        #replace no data with blank

        df.replace(np.nan, '', inplace = True)
        df.replace(pd.NaT, '', inplace = True)
        
        try:
            #Make Year of entry non decimal
            df['Intended Year of Entry'] = df['Intended Year of Entry'].fillna(0.0).astype(int)
        except:
            output_text('Something strange happened, speak to jack')
        
        createTree()
        output_text('Check the data in the preview box. If this looks right, click Save')
        
        #save df to list
        global dataList
        global fields
        dataList = df.values.tolist()
        fields = df.columns.values.tolist()
    
    except:
        output_text("This didn't seem to work. Check you have selected the correct data and try again")
```

```python
def exe():
    try:
        #Create new table 
       
        global df
        df = df[['Event Name','First Name', 'Last Name', 'Email address', 'Date of Birth','Address Country Code',
                    'Nationality code', 'MKT - Study Type',
               'MKT Entry year', 'email_consent', 'MKT - PG Subject Area']]
        
        #format DOB
        df["Date of Birth"]= pd.to_datetime(df["Date of Birth"])
        df['DOB'] = df["Date of Birth"].dt.strftime('%m/%d/%Y')
        

        df = df[['First Name', 'Last Name', 'Email address', 'DOB','Address Country Code',
                    'Nationality code', 'MKT - Study Type',
               'MKT Entry year', 'email_consent', 'MKT - PG Subject Area', 'Event Name']]
        

            # clean data and format table

        #New headers
        df = df.rename(columns = {'Nationality code': 'Standard Country',
                        'MKT - Study Type': 'Prospectus Type/Level of Study',
                         'Address Country Code':'Standard Country',
                          'email_consent': 'GDPR Consent',
                            'MKT - PG Subject Area': 'PG Subject of interest 1',
                            'Event Name': 'Event name and date',
                        'MKT Entry year': 'Entry year'})

        #Add data to Audience Type
        df.loc[(df['GDPR Consent'].isnull()), 'GDPR Consent']  = 'Email'

        #remove rows with no data

        df.dropna(axis= 'index', how='any', subset = ['Last Name','Email address'], inplace = True)

        #replace no data with blank

        df.replace(np.nan, '', inplace = True)
        df.replace(pd.NaT, '', inplace = True)
        
        try:
            #Make Year of entry non decimal
            df['Entry year'] = df['Entry year'].fillna(0.0).astype(int)
        except:
            output_text('Something strange happened, speak to jack')
        
        createTree()
        output_text('Check the data in the preview box. If this looks right, click Save')
        

        #save df to list
        global dataList
        global fields
        dataList = df.values.tolist()
        fields = df.columns.values.tolist()
        
    
    except:
        output_text("This didn't seem to work. Check you have selected the correct data and try again")
```

```python
#function to choose which execution type
def exeButton():
    if importType.get() == 1:
        execute()
        
    elif importType.get() == 2:
        exe()
    else:
        output_text("This didn't seem to work. Speak to Jack")
        
```

```python
#create the preview screen using Treeview
def createTree():
    clearTree()
    my_tree['column'] = list(df.columns)
    my_tree['show'] = 'headings'

    for column in my_tree['column']:
        my_tree.heading(column, text = column)

    df_rows = df.to_numpy().tolist()

    for row in df_rows:
        my_tree.insert('', END, values = row)
```

```python
#function to save the new csv file
def save(columns, rows):
    
    try: 
        saveFile  = filedialog.asksaveasfilename(filetypes=[('CSV File','.csv')], defaultextension = '.csv')
        with open (saveFile, mode = 'w', newline='', encoding='utf-8-sig') as file:
            csvWriter = csv.writer(file, delimiter = ',')
            csvWriter.writerow(columns)
            csvWriter.writerows(rows) 
        output_text("Your new spreadsheet has been saved")
    except:
        output_text("This didn't seem to work. Try again")
```

```python
def helpWindow():
    helpbox = Toplevel()
    helpbox.configure(background='#00a39b')
    helpbox.title("Jack's Import Converter V4 - Help")
    helpLabel = Label(helpbox, text=''' HELP: 
    - The preview window will display a preview of the data currently being held.
    - You need to select a type of import, either UG or PG then select the file using the Select file button.
    - Once you have done that a preview of the file will appear. If this looks correct click the execute button
    - which will convert it do it so it is in the correct format to be uploaded. The preview window will show this.
    - If this looks correct, click the save button and save the file. 
    - It will save as a csv which can be imported.
    - If this doesnt work ask Jack.''',justify= LEFT)
    helpLabel.pack()
```

```python
#create the GUI
root = Tk() 
root.geometry("400x600")
root.title("Jack's Import Converter V4")
root.configure(background='#00a39b')

my_frame = LabelFrame(root, text = 'Preview Window')
my_frame.config(bg = "#00a39b")
my_frame.pack(pady = 20)


x_scroll = Scrollbar(my_frame, orient = 'horizontal')
y_scroll = Scrollbar(my_frame, orient = 'vertical')

my_tree = ttk.Treeview(my_frame, xscrollcommand = x_scroll.set, yscrollcommand = y_scroll.set)

x_scroll.config(command = my_tree.xview)
y_scroll.config(command = my_tree.yview)
x_scroll.config(bg= "#004d49")
y_scroll.config(bg= "#004d49")

x_scroll.pack(side= BOTTOM, fill = X)
y_scroll.pack(side= RIGHT, fill = Y)

my_tree.pack()

my_label =Label(root, text = '''Select the type of import using the radio buttons. \n
Then select the spreadsheet using the 'Select file' button''', justify= LEFT )
my_label.config(bg= "#00a39b", fg= "white")
my_label.pack()
my_label.place(relx = 0.5, rely=0.6, anchor = 'center')


fileSelect = Button(root, text = "Select file", width = 10, padx = 10, command = lambda: fileOpen())
fileSelect.config(bg= "#004d49", fg= "white")
fileSelect.pack()
fileSelect.place(relx = 0.25, rely=0.9, anchor = 'center')

uploadChange = Button(root, text = "Execute", width = 10, padx = 10, command = lambda: exeButton())
uploadChange.config(bg= "#004d49", fg= "white")
uploadChange.pack()
uploadChange.place(relx = 0.5, rely=0.9, anchor = 'center' )

saveData = Button(root, text ="Save", width = 10,padx = 10, command = lambda: save(fields, dataList))
saveData.config(bg= "#004d49", fg= "white")
saveData.pack()
saveData.place(relx = 0.75, rely=0.9, anchor = 'center')

global importType
importType = IntVar()

radioOne =Radiobutton(root, text = "UG (HubSpot)", variable = importType, value = 1)
radioTwo =Radiobutton(root, text = "PG (Hobsons)", variable = importType, value = 2)
radioOne.config(bg= "#00a39b", fg= "white",selectcolor='#00706b')
radioTwo.config(bg= "#00a39b", fg= "white",selectcolor='#00706b')
radioOne.pack()
radioTwo.pack()
radioOne.place(relx = 0.5, rely=0.7, anchor = 'center')
radioTwo.place(relx = 0.5, rely=0.75, anchor = 'center')

helpButton =Button(root, text = 'Help', width = 10, padx = 10, command = lambda: helpWindow())
helpButton.config(bg= "#004d49", fg= "white")
helpButton.pack()
helpButton.place(relx = 0.5, rely=0.52, anchor = 'center')
```
